```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
```

```python
df_report=pd.read_csv('student_marks.csv')  
df_report.head()
```

```python
df_report['Performance']=df_report['Maths']+df_report['Physics']+df_report['Chemistry']+df_report['English']+df_report['Biology']+df_report['Economics']+df_report['History']+df_report['Civics']
```

```python
df_report.head()
```

```python
df_report["PerformancePercentile"]=(df_report['Performance']/800)*100
```

```python
df_report_cor=df_report.corr()                          #To check the correlation
df_report_cor
```

```python
plt.figure(figsize=(9,7))
sns.heatmap(df_report_cor,cmap='Reds',annot=True)
```

```python
from scipy.stats import zscore
z_score=abs(zscore(df_report.iloc[:,3:13]))
print(df_report.iloc[:,3:13].shape)
df_report_final=df_report.iloc[:,3:13].loc[(z_score<3).all(axis=1)]                  
print(df_report_final.shape)
#to quantify the outliers
```

```python
col_list=df_report.iloc[:,3:13].columns.values
plt.figure(figsize=(20,50))
for i in range(0, len(col_list)):                                       #Distplot is used To check the skewness
    plt.subplot(8,6,i+1)
    sns.distplot(df_report[col_list[i]])
```
